```python
%%sql
FROM ChicagoPublicSchools
WHERE "Elementary, Middle, or High School" = 'ES';
```

```python
%sql SELECT MAX(SAFETY_SCORE) AS MAX_SAFETY_SCORE FROM ChicagoPublicSchools
```

```python
%sql SELECT NAME_OF_SCHOOL AS SAFEST_SCHOOL FROM ChicagoPublicSchools WHERE SAFETY_SCORE= 99
```

```python
%sql SELECT NAME_OF_SCHOOL , SAFETY_SCORE FROM ChicagoPublicSchools WHERE SAFETY_SCORE=(SELECT MAX(SAFETY_SCORE) FROM ChicagoPublicSchools)
```

```python
%sql SELECT NAME_OF_SCHOOL FROM ChicagoPublicSchools ORDER BY AVERAGE_STUDENT_ATTENDANCE DESC LIMIT 10
```

```python
%sql SELECT NAME_OF_SCHOOL, AVERAGE_STUDENT_ATTENDANCE FROM ChicagoPublicSchools ORDER BY  AVERAGE_STUDENT_ATTENDANCE LIMIT 5
```

```python
%sql SELECT NAME_OF_SCHOOL, REPLACE(AVERAGE_STUDENT_ATTENDANCE,'%','') FROM ChicagoPublicSchools ORDER BY  AVERAGE_STUDENT_ATTENDANCE LIMIT 5
```

```python
%sql SELECT NAME_OF_SCHOOL, AVERAGE_STUDENT_ATTENDANCE FROM ChicagoPublicSchools WHERE CAST (REPLACE(AVERAGE_STUDENT_ATTENDANCE,'%','') AS DOUBLE) < 70 ORDER BY AVERAGE_STUDENT_ATTENDANCE
```

```python
%sql select Community_Area_Name, sum(College_Enrollment) AS TOTAL_ENROLLMENT \
   from ChicagoPublicSchools \
   group by Community_Area_Name
```

```python
%sql SELECT Community_Area_Name, sum(College_Enrollment) FROM ChicagoPublicSchools GROUP BY Community_Area_Name  ORDER BY SUM(College_Enrollment) LIMIT 5
```

```python
%sql SELECT NAME_OF_SCHOOL, SAFETY_SCORE FROM ChicagoPublicSchools WHERE SAFETY_SCORE !='None' ORDER BY SAFETY_SCORE LIMIT 5
```
